```python
import numpy as np
from scipy.sparse.linalg import eigsh
import scipy.sparse as sparse
from scipy.linalg import eigh
import scipy.linalg
import matplotlib.pyplot as plt
from qutip import *
import math
```

```python
#plt.rcParams['figure.figsize']=8.5,5.1
plt.rcParams['font.size']=20
plt.rcParams['lines.linewidth']=1.5
plt.rcParams['text.usetex']=True
```

```python
T1=1100*1/0.083
T2=210*1/0.083

N_time=150

Delta_t_1=T1/N_time
Delta_t_2=T2/N_time

t_array_1100=np.linspace(Delta_t_1,T1,N_time)
t_array_210=np.linspace(Delta_t_2,T2,N_time)

T_EVOL_T1100INVGAP_Nt150_NG_M20N50=np.load('T_EVOL_T1100INVGAP_Nt150_NG_M20N50.npy')
T_EVOL_T210INVGAP_Nt150_PHIEXT_M20N50=np.load('T_EVOL_T210INVGAP_Nt150_PHIEXT_M20N50.npy')
```

```python
# Create the figure and subplots
fig = plt.figure(figsize=(13, 5))

# Plot 1: 3D surface plot
ax1 = fig.add_subplot(121)
ax1.plot(t_array_1100,T_EVOL_T1100INVGAP_Nt150_NG_M20N50)
ax1.set_ylim(0,1)
ax1.set_xlim(0,13253)
ax1.set_xlabel(r"$T$")
ax1.set_ylabel(r"$|\langle 1_{exp} | 1_{obt} \rangle |^2$")
ax1.text(6000,0.5,r"$\mathbf{n_g}$",fontweight="bold")
ax1.text(400,0.05,r"$ | 0 \rangle $")
ax1.text(9150,0.9,r"$ | 1 \rangle $")

# Plot 1: 3D surface plot
ax2 = fig.add_subplot(122)
ax2.plot(t_array_210,T_EVOL_T210INVGAP_Nt150_PHIEXT_M20N50)
ax2.set_ylim(0,1)
ax2.set_xlim(0,2530)
ax2.set_xlabel(r"$T$")
ax2.text(960,0.5,r"$\mathbf{\varphi_{ext}}$",fontweight="bold")
ax2.text(70,0.05,r"$ | 0 \rangle $")
ax2.text(2250,0.9,r"$ | 1 \rangle $")

# Adjust the spacing between subplots
fig.subplots_adjust(wspace=0.2)

# Show the figure
#plt.savefig('Time_evolution_ng&phiext_M20N50.png',dpi=300,bbox_inches='tight')
plt.show()
```
