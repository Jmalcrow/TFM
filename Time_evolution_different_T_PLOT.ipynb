```python
#1

T1=5.5*1/0.00195092
Nt1=30
t_array_1=np.linspace(T1/Nt1,T1,Nt1)

T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_1=np.load('T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_1.npy')
T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_1=np.load('T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_1.npy')

#2

T2=5.5*1/0.0029263
Nt2=30
t_array_2=np.linspace(T2/Nt2,T2,Nt2)

T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_2=np.load('T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_2.npy')
T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_2=np.load('T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_2.npy')

#3

T3=5.5*1/0.00097546
Nt3=30
t_array_3=np.linspace(T3/Nt3,T3,Nt3)

T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_3=np.load('T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_3.npy')
T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_3=np.load('T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_3.npy')
```

```python
# Create the figure and subplots
fig = plt.figure(figsize=(18, 4))

# Plot 1
ax1 = fig.add_subplot(131)
ax1.plot(t_array_1,T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_1, label=r"$\epsilon_{n} = 0.500 \% $", c="yellowgreen")
ax1.plot(t_array_1,T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_1,label=r"$\epsilon_{\varphi_{ext}} = 1.139 \% $", c="darkblue", ls='dotted')
ax1.axhline(1, color='black', linestyle='--')
ax1.axhline(0, color='black', linestyle='--')
ax1.set_xlabel(r"$N_t$")
ax1.set_ylabel(r"$| \langle \Psi_t | 1 \rangle |^2$")
ax1.text(-600,1.3,r"$(a)$")
ax1.legend(bbox_to_anchor=(0.15,1.0),ncol=1)

# Plot 2
ax2 = fig.add_subplot(132)
ax2.plot(t_array_2,T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_2,label=r"$\epsilon_{n} = 1.000 \% $", c="yellowgreen")
ax2.plot(t_array_2,T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_2,label=r"$\epsilon_{\varphi_{ext}} = 2.278 \% $", c="darkblue", ls='dotted') 
ax2.axhline(1, color='black', linestyle='--')
ax2.axhline(0, color='black', linestyle='--')
ax2.set_xlabel(r"$N_t$")
ax2.text(-300,1.3,r"$(b)$")
ax2.legend(bbox_to_anchor=(0.15,1.0),ncol=1)

# Plot 3
ax3 = fig.add_subplot(133)
ax3.plot(t_array_3,T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_3,label=r"$\epsilon_{n} = 1.500 \% $", c="yellowgreen")
ax3.plot(t_array_3,T_EVOL_T5_5_INVGAP_Nt30_PHIEXT_M20N50_EG_3,label=r"$\epsilon_{\varphi_{ext}} = 3.417 \% $", c="darkblue", ls='dotted') 
ax3.axhline(1, color='black', linestyle='--')
ax3.axhline(0, color='black', linestyle='--')
ax3.set_xlabel(r"$N_t$")
ax3.text(-900,1.3,r"$(c)$")
ax3.legend(bbox_to_anchor=(0.15,1.0),ncol=1)

#plt.legend(bbox_to_anchor=(0.6,1.3),ncol=3)
# Adjust the spacing between subplots
fig.subplots_adjust(wspace=0.2)

# Show the figure
#plt.savefig('T_EVOL_T5_5_INVGAP_Nt30_NG_M20N50_EG_all.png',dpi=300,bbox_inches='tight')
plt.show()
```
